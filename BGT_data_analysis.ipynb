```python
a=pd.read_csv("data.csv",sep=",")
```

```python
a
```

```python
pol="GHG"
a[(a["Country"]=="India")& (a["Year"]==2000) & (a["VAR"]=="TOTAL") & (a["POL"]==pol)]["Value"]
```

```python
a[(a["Country"]=="United States")& (a["Year"]==2000) & (a["VAR"]=="TOTAL") & (a["POL"]==pol)]["Value"].sum()
```

```python
countries=list(set(a["Country"]))
```

```python
years=list(set(a["Year"]))
```

```python
c02_consumption=[]
country=[]
year=[]
for x in countries:
    for y in years:
        b=a[(a["Country"]==x)& (a["Year"]==y) & (a["VAR"]=="TOTAL") & (a["POL"]==pol)]["Value"]
        c02_consumption.append(b.sum())
        country.append(x)
        year.append(y)
```

```python
df=pd.DataFrame()
```

```python
df["Co2_consumption"]=c02_consumption
```

```python
df["year"]=year
df["country"]=country
```

```python
df.head()
```

```python
#Compute per year maximum
import numpy as np
max_in_years=[]
for x in years:
    y=np.max(df[df["year"]==x]["Co2_consumption"])
    max_in_years.append(y)
```

```python
max_in_years
```

```python
years
```

```python
import matplotlib.pyplot as plt
plt.plot(years,max_in_years)
plt.xlabel('years')
plt.ylabel('Max per country CO2 emissions')
```

```python
#Compute per year maximum
import numpy as np
mean_in_years=[]
for x in years:
    y=np.mean(df[df["year"]==x]["Co2_consumption"])
    mean_in_years.append(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(years,mean_in_years)
plt.xlabel('years')
plt.ylabel('Mean GHG emissions')
```

```python
plt.hist(means)
plt.xlabel('Mean Co2 emmisions per country')
plt.ylabel('Number of countries')
```
